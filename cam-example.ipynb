```python
from cam import get_superimposed_batch, get_batch, get_heatmaps
import tensorflow as tf
import os
```

The script extracts Class Activation Maps batch from the batch of images using Tensorflow methods

```python
"""
ImageNet classes:
207 - golden retriever
251 - dalmatian, coach dog, carriage dog
"""

model = tf.keras.applications.ResNet50V2()
input_size = (224, 224)

# Path pattern can be the glob pattern or the list of paths
img_dir = 'images'
# img_type = 'jpg'
# path_pattern = os.path.join(img_dir, f'*.{img_type}')
path_pattern = ['Golden_Retriever_with_tennis_ball.jpg', '200px-Dalmatian_puppy,_four_months.png']
path_pattern = list(map(lambda path: os.path.join(img_dir, path), path_pattern))

# Class index or class index for each image (list of indices) can be used as the argument to get heatmaps corresponding the class output
# If class_index is not defined, classes with max preds for each image will be used to get heatmaps
class_index = [207, 251]

# The images with heatmaps can be saved into the predefined directory (save_dir)
# save_superimposed_batch(img_dir, img_type, save_dir, model = model, class_index = [282, 208])

batch = get_batch(path_pattern, img_size = input_size)
heatmaps = get_heatmaps(batch, model, class_index = class_index)
superimposed_batch = get_superimposed_batch(batch, heatmaps, alpha=0.4)
```

```python
import matplotlib.pyplot as plt

plt.imshow(superimposed_batch[0])
plt.show()
```
